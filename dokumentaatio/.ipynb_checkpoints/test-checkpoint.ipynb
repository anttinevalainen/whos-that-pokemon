```python
image = cv2.imread('kuvat/pokemon/700.png')
mask = np.ones(image.shape, dtype=np.uint8) * 255
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cnts = cv2.findContours(gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
for c in cnts:
    cv2.drawContours(mask, [c], -1, (0, 0, 0), thickness=1)
    
img = Image.fromarray(mask, 'RGB')
img = img.convert("RGBA")
pixdata = img.load()

width, height = img.size
for y in range(height):
    for x in range(width):
        if pixdata[x, y] == (255, 255, 255, 255):
            pixdata[x, y] = (255, 255, 255, 0)

width, height = img.size
center = (int(0.5 * width), int(0.5 * height))
yellow = (0, 0, 0, 255)
ImageDraw.floodfill(img, xy=center, value=yellow)

img
```

```python
# Convert to grayscale
mask = Image.open('kuvat/pokemon/615.png').convert('L')
mask = mask.point(lambda i: i < 1 and 255)
# Threshold and invert the colors (white will be transparent)
mask = mask.point(lambda x: x < 100 and 255)

# The size of the images must match before apply the mask
img = Image.open('kuvat/pokemon/black.png')

img.putalpha(mask) # Modifies the original image without return

img

width, height = img.size
for x in range(width):
    for y in range(height):
        
        top = (x,y-1)
        bottom = (x,y+1)
        left = (x-1,y)
        right = (x+1,y)
        leftover = (x-1,y-1)
        rightover = (x+1,y-1)
        leftunder = (x-1,y+1)
        rightunder = (x+1,y+1)
        current = (x,y)
        i = 0
        
        if img.getpixel(current) == (0,0,0,255):
            continue
        elif y == 255 or x == 255 or y == 0 or x == 0:
            continue
        else:
            if img.getpixel(top) == (0,0,0,255):
                i+=1
            if img.getpixel(bottom) == (0,0,0,255):
                i+=1
            if img.getpixel(left) == (0,0,0,255):
                i+=1
            if img.getpixel(right) == (0,0,0,255):
                i+=1
            if img.getpixel(leftover) == (0,0,0,255):
                i+=1
            if img.getpixel(leftunder) == (0,0,0,255):
                i+=1
            if img.getpixel(rightover) == (0,0,0,255):
                i+=1
            if img.getpixel(rightunder) == (0,0,0,255):
                i+=1
                
            if i > 4:
                img.putpixel((current), (0,0,0,255))
img
```
